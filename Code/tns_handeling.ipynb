```python
import pandas as pd
from astropy.table import Table
```

```python
#load tns data
tns_table = Table.read("../Data/tns_public_object.csv", header_start=1, format="csv")
print(len(tns_table))
```

```python
#mask out non-SN
tns_sn_mask = tns_table["name_prefix"] == "SN"
tns_sn = tns_table[tns_sn_mask]
```

```python
#save tns SN data
tns_sn.write("../Data/tns_sn.csv", format="ascii.csv")
```
