```python
import pandas as pd
from astropy.table import Table
import numpy as np
```

```python
#load tns data
tns_table = Table.read("../Data/tns_public_object.csv", header_start=1, format="csv")
print(len(tns_table))
```

```python
#mask out non-SN
tns_sn_mask = tns_table["name_prefix"] == "SN"
tns_sn = tns_table[tns_sn_mask]
```

```python
#save tns SN data
tns_sn.write("../Data/tns_sn.csv", format="ascii.csv")
```

```python
#create tables of only specific sn types and ZTF or Rubin eg. Ia, Ib, Ic, II 

#Ia
mask_Ia = (tns_sn["type"] == "SN Ia") & (np.logical_or((tns_sn["source_group"] == "ZTF"), (tns_sn["source_group"] == "Rubin")))

#Ib
mask_Ib = (tns_sn["type"] == "SN Ib") & (np.logical_or((tns_sn["source_group"] == "ZTF"), (tns_sn["source_group"] == "Rubin")))

#Ic
mask_Ic = (tns_sn["type"] == "SN Ic") & (np.logical_or((tns_sn["source_group"] == "ZTF"), (tns_sn["source_group"] == "Rubin")))

#II
mask_II = (tns_sn["type"] == "SN II") & (np.logical_or((tns_sn["source_group"] == "ZTF"), (tns_sn["source_group"] == "Rubin")))
```

```python
#mask tns sn for each type
tns_Ia = tns_sn[mask_Ia]
print(f"#Ia: {len(tns_Ia)}")

tns_Ib = tns_sn[mask_Ib]
print(f"#Ib: {len(tns_Ib)}")

tns_Ic = tns_sn[mask_Ic]
print(f"#Ic: {len(tns_Ic)}")

tns_II = tns_sn[mask_II]
print(f"#II: {len(tns_II)}")
```

```python
#save tns tables
tns_Ia.write("../Data/tns_Ia.csv", format="ascii.csv")
tns_Ib.write("../Data/tns_Ib.csv", format="ascii.csv")
tns_Ic.write("../Data/tns_Ic.csv", format="ascii.csv")
tns_II.write("../Data/tns_II.csv", format="ascii.csv")
```
